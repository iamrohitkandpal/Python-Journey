```python
df = pd.read_csv(r'anime.csv')
df.head()
```

```python
df.loc[2]["Title"]
```

```python
def extract_episodes(txt):
    check = False 
    data = ""
    for i in txt:
        if i == ")":
            check = "False"
            return data
        if check == True:
            data = data + i
        if i == "(":
            check = True
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
df["Episodes"] = df["Episodes"].str.replace(" eps", "")
df["Episodes"] = df["Episodes"].astype(int)
df.head()
```

```python
def extract_stamps(txt):
    check = False
    count = 0
    data = ""
    for i in txt:
        if count > 18:
            check = "False"
            count = 0
            return data
        if check == True:
            data = data + i
            count += 1
        if i == ")":
            check = True
```

```python
df["Tenure"] = df["Title"].apply(extract_stamps)
df.head()
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def extract_months(txt):
    try:
        start, end = txt.split(' - ')
        start_date = datetime.strptime(start, "%b %Y")
        end_date = datetime.strptime(end, "%b %Y")
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1 
    except:
        return None

df["Months"] = df["Tenure"].apply(extract_months)
df.head()
```
