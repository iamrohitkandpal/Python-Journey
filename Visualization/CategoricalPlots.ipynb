```python
import seaborn as sns
import numpy as np
```

***************************************
**<h1>Categorical Data Plots</h1>**

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

Let's go through examples of each!
***************************************

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(x = df['sex'], hue = df['smoker'])
```

```python
sns.barplot(x = df['sex'], y = df['total_bill'])
```

```python
sns.barplot(x = df['sex'], y = df['tip'], estimator = np.median)
```

```python
sns.boxplot(x = 'tip', y = 'day', data = df, hue = 'size', palette = 'rainbow')
```

```python
sns.violinplot(x = 'total_bill', y = 'sex', hue = 'smoker', data = df)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = df)
sns.swarmplot(x = 'day', y = 'total_bill', data = df)
```
