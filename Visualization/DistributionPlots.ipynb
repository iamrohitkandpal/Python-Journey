```python
import seaborn as sns
import matplotlib.pyplot as plt
```

***********************************
**We use Distribution plots to visualise a distribution of quantitative data** 

**Let's discuss some plots that allow us to visualise the distribution of our data set. These plots are:**
- displot/histplot
- jointplot
- pairplot
- rugplot
************************************

```python
df = sns.load_dataset("tips")
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=20, kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins=20, kde = True)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
sns.pairplot(df, hue='size', palette='rainbow')
```

```python
sns.rugplot(df)
```
