```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

*************************
**Matplotlib Basics**
*********************

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.title("Company Report")
plt.xlabel("Months")
plt.ylabel("Sales")
```

********************
**Subplots in Matplotlib**
***********************

```python
plt.subplot(2,2,1)
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(y,y)
```

****************************
**Object Oreinted Way**
**************************

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You use the figsize and dpi keyword arguments. 

figsize is a tuple of the width and height of the figure in inches
dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axis1.plot(x,y)
axis2 = fig.add_axes([0.25, 0.6, 0.3, 0.3])
axis2.plot(x,y)
```
