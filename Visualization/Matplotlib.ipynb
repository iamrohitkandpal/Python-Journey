```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

*************************
**Matplotlib Basics**
*********************

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.title("Company Report")
plt.xlabel("Months")
plt.ylabel("Sales")
```

********************
**Subplots in Matplotlib**
***********************

```python
plt.subplot(2,2,1)
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(y,y)
```
