```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

*****************************
**<h2>lmplot()</h2>**

*Regression Plot is used in machine learning, so we will not fully discuss what, how and where it is used. But we will see how it is plated and what are the values we can put inside it.*|
**************************************

```python
df = sns.load_dataset('tips')
df
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', data = df, palette = 'coolwarm')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', data = df, palette = 'rainbow', markers = ['o', 'v'], scatter_kws = {'s': 50})
```
